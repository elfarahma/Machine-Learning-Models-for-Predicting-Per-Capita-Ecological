```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
from sklearn.preprocessing import normalize
```

```python
data = pd.read_csv('data_for_plotly_grid.csv')
data
```

```python
#boxplot continent vs HDI
sns.boxplot(x='continent', y='hdi', data=data, )

plt.show()
```

```python
#boxplot continent vs EFConsPerCap
sns.boxplot(x='continent', y='EFConsPerCap', data=data)
plt.show()
```

```python
#plot regresi year vs EFConsPerCap

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
sns.regplot(x='year',y='EFConsPerCap',data=data, ax=axes[0] )
sns.regplot(x='year',y='hdi',data=data, ax=axes[1] )
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Ecological Footprint per Capita (GHA)")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Human Development Index (HDI)")
plt.show()
```

```python
#plot regresi antara HDI dan EF per kapita

ax = sns.lmplot( x="hdi", y="EFConsPerCap", data=data, fit_reg=False, 
           hue='continent', height=6, aspect=1.5, scatter_kws={"alpha": 0.5},
           legend=False)
 
plt.legend(loc='upper left')
plt.xlabel("Human Development Index (HDI)")
plt.ylabel("Ecological Footprint per Capita (GHA)")
plt.show()
```

```python
x = np.log10(data["SalePrice"])
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```
